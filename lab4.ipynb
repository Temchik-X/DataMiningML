### Создание генераторов

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

image.png

```python
train = train_datagen.flow_from_directory(
    directory=os.path.join(root, train_folder_name),
    batch_size=32,
    target_size=(256, 256),
    class_mode='categorical'
)
validation = val_datagen.flow_from_directory(
    directory=os.path.join(root, validation_folder_name),
    batch_size=32,
    target_size=(256, 256),
    class_mode='categorical'
)
test = test_datagen.flow_from_directory(
    directory=os.path.join(root, test_folder_name),
    batch_size=32,
    target_size=(256, 256),
    class_mode='categorical'
)
```

## Задание №1

```python
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(shape=(256, 256, 3)), # Высота, ширина, количество цветовых каналов
    tf.keras.layers.Conv2D(32, (3, 3), # Размер фильтра, указывает на область (ширину × высоту), которая используется для извлечения признаков.
                                        activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),# Указывает, какая область обрабатывается для уменьшения размерности.
    tf.keras.layers.Conv2D(64, (3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(256, (3, 3), activation="relu"),
    # Dropout — это регуляризационный слой, который помогает предотвратить переобучение, отключая случайно выбранные нейроны во время обучения.
    tf.keras.layers.Dropout(0.3),
    # Преобразует многомерный массив (например, выход из свёрточных слоёв) в одномерный вектор. Это необходимо для подключения к Dense слоям
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation="relu"),
    tf.keras.layers.Dense(5, activation="softmax"),
])

model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss='categorical_crossentropy',
    metrics=["accuracy"],
)
```

```python

callbacks = [
    # Колбэк EarlyStopping: останавливает обучение, если "val_accuracy" не улучшается в течение 10 эпох
    tf.keras.callbacks.EarlyStopping(
        monitor="val_accuracy",   # Мониторим метрику "val_accuracy" на проверочной выборке
        mode="max",               # Ожидается максимизация значения (лучшее значение)
        patience=10               # Количество эпох без улучшения перед остановкой
    ),
    # Колбэк ModelCheckpoint: сохраняет лучшую модель по метрике "val_accuracy"
    tf.keras.callbacks.ModelCheckpoint(
        filepath="models/best_model_conv.keras",  # Путь для сохранения модели
        monitor="val_accuracy",                  # Мониторим "val_accuracy"
        mode="max",                              # Сохраняем модель с максимальным значением
        save_best_only=True                      # Сохраняем только лучшую модель
    ),
]
```

###Краткое описание модели
**Модель:**<br>
Многослойная свёрточная нейронная сеть с Conv2D, MaxPooling2D, Dropout, Flatten и Dense слоями.<br>
 **Dropout** — это регуляризационный слой, который помогает предотвратить переобучение, отключая случайно выбранные нейроны во время обучения.<br>
 **Flatten** - преобразует многомерный массив (например, выход из свёрточных слоёв) в одномерный вектор. Это необходимо для подключения к Dense слоям

**Тренировка:**<br>
Выполняется на обучающих данных (train), с проверкой на валидационных данных (validation).
100 шагов за эпоху для обучения и 50 шагов для валидации.
Максимум 100 эпох, но обучение может завершиться раньше благодаря EarlyStopping.

**Колбэки:**<br>
EarlyStopping: Прекращает обучение, если val_accuracy не улучшается в течение 10 эпох.
ModelCheckpoint: Сохраняет лучшую модель на основе максимального значения val_accuracy.

```python
# Обучение модели
history = model.fit(
    train,                    # Обучающая выборка (генератор)
    steps_per_epoch=100,      # Количество шагов за одну эпоху для train
    validation_data=validation,  # Проверочная выборка (генератор)
    validation_steps=50,      # Количество шагов для проверки за эпоху
    epochs=100,               # Максимальное количество эпох
    callbacks=callbacks
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend(loc='lower right')

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend(loc='upper right')

plt.show()
```

```python
loss, accuracy = best_model_conv.evaluate(test)
print(f"Loss: {loss:.3f}")
print(f"Accuracy: {accuracy:.3f}")
```

## Задание №2

### Автокодировщик для предобучения


Приводим данные к размеру

```python
# Загрузка данных
train_autoencoder = tf.keras.utils.image_dataset_from_directory(
    directory=os.path.join(root, train_folder_name),
    batch_size=32,
    image_size=(256, 256)
)

validation_autoencoder = tf.keras.utils.image_dataset_from_directory(
    directory=os.path.join(root, validation_folder_name),
    batch_size=32,
    image_size=(256, 256)
)

# Подготовка данных для автокодировщика
train_autoencoder = train_autoencoder.map(lambda x, y: (x / 255.0, x / 255.0))
validation_autoencoder = validation_autoencoder.map(lambda x, y: (x / 255.0, x / 255.0))
```

Для текущей задачи лучше подходит сверточный автокодировщик, потому что он эффективно обрабатывает изображения, извлекая пространственные и цветовые признаки, важные для распознавания цветов, таких как тюльпаны или розы. Сверточные слои хорошо подходят для анализа текстур, форм и цветовых оттенков, сохраняя локальные зависимости в данных.

В отличие от обычных автокодировщиков с полносвязными слоями, сверточный автокодировщик значительно снижает количество параметров, что делает его более вычислительно эффективным и позволяет работать с большими изображениями без потери пространственной структуры.

```python
# Энкодер
conv_encoder = tf.keras.models.Sequential([
    tf.keras.layers.InputLayer(shape=(128, 128, 3)),
    tf.keras.layers.Conv2D(32, kernel_size=3, strides=2, activation='selu', padding='same'),
    tf.keras.layers.Conv2D(64, kernel_size=3, strides=2, activation='selu', padding='same'),
    tf.keras.layers.Conv2D(128, kernel_size=3, strides=2, activation='selu', padding='same'),
    tf.keras.layers.Conv2D(256, kernel_size=3, strides=2, activation='selu', padding='same'),
])

# Декодер
conv_decoder = tf.keras.models.Sequential([
    tf.keras.layers.Conv2DTranspose(128, kernel_size=3, strides=2, activation='selu', padding='same'),
    tf.keras.layers.Conv2DTranspose(64, kernel_size=3, strides=2, activation='selu', padding='same'),
    tf.keras.layers.Conv2DTranspose(32, kernel_size=3, strides=2, activation='selu', padding='same'),
    tf.keras.layers.Conv2DTranspose(16, kernel_size=3, strides=2, activation='selu', padding='same'),
    tf.keras.layers.Conv2D(3, kernel_size=3, activation='sigmoid', padding='same')  # 3 канала RGB
])

# Автокодировщик
conv_autoencoder = tf.keras.models.Sequential([conv_encoder, conv_decoder])

conv_autoencoder.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
conv_autoencoder.summary()
```

**Модель:**
Автокодировщик, состоящий из сверточного энкодера и декодера.

**Энкодер:**<br>
Четыре последовательных слоя Conv2D с шагом 2 и активацией SELU. Каждый слой увеличивает количество фильтров (32 → 64 → 128 → 256), сжимая входное изображение размером (128, 128, 3) до латентного представления размером (8, 8, 256).

**Декодер:**<br>
Четыре слоя Conv2DTranspose с шагом 2 и активацией SELU. Декодер восстанавливает исходные размеры изображения, постепенно уменьшая количество фильтров (128 → 64 → 32 → 16). Последний слой Conv2D с активацией sigmoid выдает восстановленное изображение размером (128, 128, 3).

**Тренировка:**<br>
Обучение выполняется на нормализованных данных (пиксели от 0 до 1) с использованием функции потерь binary_crossentropy и оптимизатора Adam. Модель обучается на обучающем наборе с валидацией на проверочном.
Процесс ограничен максимум 100 эпохами с досрочной остановкой, если метрика валидации перестает улучшаться.

```python
callbacks = [
    # Остановка обучения, если качество на валидации перестало улучшаться
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=10),

    # Сохранение лучшей модели по результатам на валидации
    tf.keras.callbacks.ModelCheckpoint(
        'models/best_autoencoder.keras',
        monitor='val_loss',
        save_best_only=True),
]

# Создаем и обучаем автокодировщик
history_ae = conv_autoencoder.fit(train_autoencoder, epochs=20, validation_data=validation_autoencoder, callbacks=callbacks)
```

```python
acc = history_ae.history['accuracy']
val_acc = history_ae.history['val_accuracy']
loss = history_ae.history['loss']
val_loss = history_ae.history['val_loss']

epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Точность на этапе обучения')
plt.plot(epochs, val_acc, 'b', label='Точность на этапе проверки')
plt.title('Точность на этапе обучения и проверки')
plt.legend(loc='lower right')

plt.figure()

plt.plot(epochs, loss, 'bo', label='Потери на этапе обучения')
plt.plot(epochs, val_loss, 'b', label='Потери на этапе проверки')
plt.title('Потери на этапе обучения и проверки')
plt.legend(loc='upper right')

plt.show()
```

Модель демонстрирует способность эффективно учиться на предоставленных данных и сохраняет высокую точность при проверке на новых данных.

Загружаем лучший автокодировщик, для дальнейшего его использования в модели

```python
# Загружаем лучшую модель
best_autoencoder = tf.keras.models.load_model('models/best_autoencoder.keras')

# Явное указание входного слоя
inputs = tf.keras.layers.Input(shape=(128, 128, 3))  # Размер входного изображения

# Применяем все слои автокодировщика, кроме последнего
x = inputs
for layer in best_autoencoder.layers[:-1]:  # Проход по слоям до декодера
    x = layer(x)

# Создаем модель кодировщика
encoder = tf.keras.models.Model(inputs=inputs, outputs=x)

# Выводим информацию о модели
encoder.summary()
```

```python
# Создаем датасет из DirectoryIterator
train = train_datagen.flow_from_directory(
    directory=os.path.join(root, train_folder_name),
    batch_size=32,
    target_size=(128, 128),
    class_mode='categorical'
)

validation = train_datagen.flow_from_directory(
    directory=os.path.join(root, validation_folder_name),
    batch_size=32,
    target_size=(128, 128),
    class_mode='categorical'
)
# Заморозка слоев кодировщика
for layer in encoder.layers:
    layer.trainable = False
# Полносвязная модель на основе кодировщика
modeFCNN = tf.keras.models.Sequential([
    encoder,  # Предобученный кодировщик
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),  # Полносвязный слой
    tf.keras.layers.Dense(64, activation='relu'),  # Полносвязный слой
    tf.keras.layers.Dropout(0.2),  # Регуляризация
    tf.keras.layers.Dense(5, activation='softmax')  # Выходной слой
])

# Компиляция модели
modeFCNN.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Проверка структуры модели
modeFCNN.summary()
```

**Модель:**<br>
Классификационная модель, построенная на основе предварительно обученного автокодировщика.

**Энкодер:** Извлечён из предварительно обученного автокодировщика, преобразует изображения размером 128x128x3 в сжатое представление. Все слои энкодера заморожены для предотвращения изменения их параметров.

**Классификатор:** Полносвязная часть с:<br>
**Слоями Dense** с активацией relu для выделения высокоуровневых признаков из сжатого представления.<br>
**Регуляризацией** с помощью Dropout, которая снижает риск переобучения.<br>
**Выходным слоем** с активацией softmax для классификации на 5 классов.

**Тренировка:**
Модель обучается на предобработанных данных:

Замороженные слои энкодера не изменяются, тренируются только параметры классификатора.

```python
callbacks = [
    # Остановка обучения, если качество на валидации перестало улучшаться
    tf.keras.callbacks.EarlyStopping(
        monitor="val_accuracy",
        mode="max",
        patience=20),
    # Сохранение лучшей модели по результатам на валидации
    tf.keras.callbacks.ModelCheckpoint(
        filepath="models/best_model_FCNN.keras",
        monitor="val_accuracy",
        mode="max",
        save_best_only=True,
    ),
]
modeFCNN_history = modeFCNN.fit(train, epochs=100, validation_data=validation, callbacks=callbacks)
```

```python
acc = modeFCNN_history.history['accuracy']
val_acc = modeFCNN_history.history['val_accuracy']
loss = modeFCNN_history.history['loss']
val_loss = modeFCNN_history.history['val_loss']

epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Точность на этапе обучения')
plt.plot(epochs, val_acc, 'b', label='Точность на этапе проверки')
plt.title('Точность на этапе обучения и проверки')
plt.legend(loc='lower right')

plt.figure()

plt.plot(epochs, loss, 'bo', label='Потери на этапе обучения')
plt.plot(epochs, val_loss, 'b', label='Потери на этапе проверки')
plt.title('Потери на этапе обучения и проверки')
plt.legend(loc='upper right')

plt.show()
```

```python
best_model_FCNN = tf.keras.models.load_model("models/best_model_FCNN.keras")
```

```python
test = train_datagen.flow_from_directory(
    directory=os.path.join(root, test_folder_name),
    batch_size=32,
    target_size=(128, 128),
    class_mode='categorical'
)
loss, accuracy = best_model_FCNN.evaluate(test)
print(f"Loss: {loss:.3f}")
print(f"Accuracy: {accuracy:.3f}")
```

На графиках отображены метрики обучения и валидации сверточной классификационной модели, построенной на основе предварительно обученного автокодировщика. <br>Точность на тренировочной выборке стабильно растет, что указывает на успешное обучение модели. Валидационная точность также демонстрирует положительный тренд, подтверждая хорошую обобщающую способность модели. <br>Потери на тренировочной выборке показывают значительное снижение, что свидетельствует о сокращении ошибок модели. Валидационные потери также уменьшаются, хотя и колеблются, что может указывать на потенциальные области для дальнейшей настройки модели. <br>В целом, модель показывает позитивные результаты и эффективно справляется с задачей классификации на обоих наборах данных.
